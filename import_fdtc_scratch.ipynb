Method 1: loop through all acquisitions in FW of certain label, check if in FTDC project, check if in excel list

```python
adding_ftdc = []
acquisitions = fw.acquisitions.find('label=t1_mpr_AX_MPRAGE')
for acquisition in acquisitions:
    subject = fw.get(acquisition.parents["subject"])
    if acquisition.parents["project"] =='5e1cec8e3a9694005b6d12ab' and subject.label in excel_sid_list:
        adding_ftdc.append(subject.label)
        #collection.add_acquisition(acquisition.id)
        
print(len(adding_ftdc))
```

Method 2: loop through sessions in FTDC, check if sub in list, loop through acqs

```python
sessions_list = []
subs_list = []
acq_list = []
for ses in project.sessions():
    sub = fw.get(ses.parents['subject'])
    fw_ses_label = ses.label.split('-')[0]
    if sub.label in excel_sid_list:
        subs_list.append(sub.label)
        print('--')
        print(sub.label)
        row = excel.loc[excel['INDDID']==float(sub.label)]
        excel_ses_label = list(row['MRIDateFormatted'])[0]
        if fw_ses_label == str(excel_ses_label):
            sessions_list.append(ses.label)
            print(fw_ses_label)
            for acq in ses.acquisitions():
#                 print(acq.files[0]['classification'])
                has_mes = 'Measurement' in acq.files[0]['classification']
                if has_mes:
                    has_t1 = 'T1' in acq.files[0]['classification']['Measurement']
                else:
                    continue
#                 print('has measurements: ' + str(has_mes))
#                 print('has T1: ' + str(has_t1))
                if has_mes and has_t1:
                    print("HASSSSSSSSSSSSSSSS")
                    print(acq.label)
                    acq_list.append(acq.label)
print("Done!")                
```
