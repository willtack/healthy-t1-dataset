## Link image file paths to correct rows in spreadsheet

```python
import pandas as pd
import os
import glob 
```

```python
## load spreadsheet
csv_path = './all_subs.csv'
df = pd.read_csv(csv_path)
```

```python
# loop through files in folder
local_dir = '/media/will/My Passport/Ubuntu/cortical_thickness_maps/all_maps/'
for file in list(glob.glob(local_dir + '*.nii.gz')):
    basename = os.path.basename(file)
    subject_label = basename.split('-')[1].split('_')[0] # extract subject label
    # find the row with that subject label in SID column and set Path value in that row to the filepath
    df.loc[df['SID']==subject_label, 'Path'] = file  
```

```python
df
```

```python
# save csv
df.to_csv(csv_path)
```

```python
# check for duplicates
import collections
path_series = df['Path']
print([item for item, count in collections.Counter(list(path_series)).items() if count > 1])
```
